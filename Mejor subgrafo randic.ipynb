## Ver propiedades de los mejores subgrafos

Se conjetura que estos subgrafos pueden tener propiedades particulares

```python
# definir cuales son las propiedades que se conjeturan
def subgrafo_cumple_propiedades_conjetura(subgrafo_checar):
    
    '''
    Devuelve True si el grafo tiene las propiedades que se conjeturan
    
    De otro modo devuelve False
    '''
    
    # tomar los componentes conexos de este subgrafo
    componentes_conexos = [subgrafo_checar.subgraph(nodos_componente_i)
                          for nodos_componente_i in nx.connected_components(subgrafo_checar)]
    
    # ver si se cumple la propiedad en cada componente
    propiedades_componentes = [nx.is_bipartite(componente) or is_regular(componente)
                              for componente in componentes_conexos]
    
    return np.all(propiedades_componentes)
```

```python
# checar en todos los grafos peque√±os

# tomar los grafos a checar
lista_grafos = obtener_grafos_conexos_peque√±os()

# guardar donde no se cumple
# guardar los pares grafo subgrafo
grafos_subgrafos_incumple = []


# iterar en cada grafo
for idx_grafo, grafo in enumerate(lista_grafos):
    
    # ver el progreso
    print(f"Progreso: [{idx_grafo}/{len(lista_grafos)}]")
    
    # obtener los subgrafos con randic maximal
    subgrafos_randic_maximal = mejores_subgrafos_randic(grafo)
    
    
    # ir iterando en estos
    for s in subgrafos_randic_maximal:
        
        # si no se cumple, decirlo
        if not subgrafo_cumple_propiedades_conjetura(s):
            
            # este subgrafo de este grafo es un contraejemplo
            print("CONTRAEJEMPLO")
            
            # agregar a la lista
            grafos_subgrafos_incumple.append((grafo, s))
            
    # end for subgrafos
    
# en for grafos
```

```python
# decir para cuantos no
print(f"Se incumple la conjetura para {len(grafos_subgrafos_incumple)} casos de subgrafos")
```

**Entonces, para todo grafo, los subgrafos con indice de randic maximal son la union disjunta de grafos regulares o bipartitas**
